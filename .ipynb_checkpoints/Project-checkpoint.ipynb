```python
import networkx as nx
import collections
import matplotlib.pyplot as plt
from networkx.generators import random_graphs
import pandas as pd
import itertools
from netwulf import visualize
from webweb import Web
import random
import numpy as np
import seaborn as sns
from networkx.algorithms.shortest_paths.generic import shortest_path
```

```python
# US Data and Plots
```

```python
data=pd.read_csv('Data/opsahl-usairport/out.opsahl-usairport',header=None)
new_data=data[0].apply(lambda x:tuple(int(i) for i in x.split(" ")))
```

```python
nodes=max(max([i[0] for i in new_data.values]),max([i[1] for i in new_data.values]))
first_node=min(min([i[0] for i in new_data.values]),min([i[1] for i in new_data.values]))
```

```python
graph=nx.DiGraph()
graph.add_nodes_from(range(1,nodes+1))
graph.add_weighted_edges_from(new_data.values)
```

```python
class Graph:
    
    def __init__(self,G=nx.Graph()):
        self.graph=G
        self.nodes=len(self.graph.nodes)
    
    def add_edge(self,edge):
        self.graph.add_edge(*edge)
    
    def add_multiple_edges(self,edges):
        self.graph.add_edges_from(edges)
        
    def add_node(self,node):
        self.graph.add_node(node)
        
    def add_multiple_nodes(self,nodes):
        self.graph.add_nodes_from(nodes)
        
    def remove_edge(self,edge):
        self.graph.remove_edge(*edge)
    
    def remove_multiple_edges(self,edges):
        self.graph.remove_edges_from(edges)
    
    def remove_node(self,node):
        self.graph.remove_edge(node)
    
    def remove_multiple_nodes(self,nodes):
        self.graph.remove_nodes_from(nodes)
        
    def average_degree(self):
        degree_sum=len(self.graph.edges())*2
        
        return degree_sum/len(self.graph.nodes)
    
    def in_degree_distribution(self):
        x=[]
        y=[]
        a=list(dict(self.graph.in_degree()).values())
        temp=list(set(a))
        temp.sort()
        f=0
        s=len(a)
        for i in temp:
            f=f+a.count(i)/s
            y.append(f)
        plt.xlabel("K")
        plt.ylabel('p_k')
        plt.title("Cumulative In Degree Distribution")
        plt.scatter(temp,y,c='r',s=[5])
        
    def out_degree_distribution(self):
        x=[]
        y=[]
        a=list(dict(self.graph.out_degree()).values())
        temp=list(set(a))
        temp.sort()
        f=0
        s=len(a)
        for i in temp:
            f=f+a.count(i)/s
            y.append(f)
        plt.xlabel("K")
        plt.ylabel('p_k')
        plt.title("Cumulative Out Degree Distribution")
        plt.scatter(temp,y,c='r',s=[5])
    
    def degree_distribution(self):
        
        x=[]
        y=[]
        a=list(dict(self.graph.degree()).values())
        temp=list(set(a))
        temp.sort()
        f=0
        s=len(a)
        for i in temp:
            f=f+a.count(i)/s
            y.append(f)
        plt.xlabel("K")
        plt.ylabel('p_k')
        plt.title("Cumulative Degree Distribution")
        plt.scatter(temp,y,c='r',s=[5])
    
    def indegree_vs_outdegree(self):
        out=list(dict(self.graph.out_degree()).values())
        ind=list(dict(self.graph.in_degree()).values())
        plt.scatter(out,ind,c='r',s=[1])
        plt.xlabel('Out Degree')
        plt.ylabel('In Degree')
        plt.title('Out Degree Vs In Degree')
        
    def average_neighbour_degree(self):
        avg_neighbour_degree=nx.average_neighbor_degree(self.graph)
        K=self.degree()
        K=[K[i] for i in range(1,len(self.get_nodes())+1)]
        C=[avg_neighbour_degree[i] for i in range(1,len(self.get_nodes())+1)]
        X=list(set(K))
        X.sort()
        Y=[]
        print("Unique values of Degree :",len(X))
        for i in X:
            temp=0
            c=0

            for j in range(len(K)):
                if K[j]==i:
                    temp+=C[j]
                    c+=1
            Y.append(temp/c)
        plt.xlabel('Degree')
        plt.ylabel('Average Neighbour Degree')
        plt.title('Average Neighbouring Degree vs Degree')
        
        plt.scatter(X,Y,c='g',s=[5])
    
    def shortest_path_length_distribution(self):
        sp=shortest_path(self.graph)
        arr=[]
        for i in range(1,self.nodes+1):
            for j in range(i+1,self.nodes+1):
                try:
                    arr.append(len(sp[i][j])-1)
                except KeyError:
                    pass
                    
        y=[]
        x=list(set(arr))
        for i in x:
            y.append(arr.count(i))
        plt.scatter(x,y,c='g',s=[20])
        plt.xlabel("Shortest Path Length")
        plt.ylabel("Frequency of flight routes")
        plt.title("Shortest Path Length Plot")
    
    def connected_components(self):
        return list(nx.connected_components(self.graph))
    
    def degree(self):
        return self.graph.degree()
    
    def average_clustering_coefficient(self):
        return nx.average_clustering(self.graph)
    
    def clustering_coefficient(self):
        return nx.clustering(self.graph)
    
    def characteristic_path_length(self):
        return nx.average_shortest_path_length(self.graph)
    
    def get_nodes(self):
        return self.graph.nodes
    
    def knn(self):
        return nx.k_nearest_neighbors(self.graph)
    def cc_distribution(self):
        cc=nx.clustering(self.graph)
        d=list(cc.values())
        d.sort()
        
        s=sum(d)
        y=[]
        f=0
        for i in d:
            f+=i/s
            y.append(f)
        plt.yticks(np.arange(0,1,step=0.1))
        plt.scatter(d,y,c='g',s=[1])

        plt.xlabel("Local Clustering Coefficient")
        plt.ylabel("Probability")
        plt.title("Clustering Coefficient Distribution")
    
    def betweeness_distribution(self):
        bc=nx.betweenness_centrality(self.graph)

        X=list((bc.values()))
        temp=list(set(X))
        temp.sort()
        Y=[]
        f=0
        s=len(X)
        for i in temp:
            f=f+X.count(i)/s
            Y.append(f)

        plt.scatter(temp,Y,c='r',s=[1])

        plt.title("Cumulative Betweeness Centrality Distribution")
        plt.xlabel('Betweeness Centrality')
        plt.ylabel("p_k")
        
    def closeness_distribution(self):
        cc=nx.closeness_centrality(self.graph)

        X=list((cc.values()))
        temp=list(set(X))
        temp.sort()
        Y=[]
        f=0
        s=len(X)
        for i in temp:
            f=f+X.count(i)/s
            Y.append(f)

        plt.scatter(temp,Y,c='r',s=[1])

        plt.title("Cumulative Betweeness Centrality Distribution")
        plt.xlabel('Betweeness Centrality')
        plt.ylabel("p_k")
        
    def betweeness_vs_nodes(self):
        bc=nx.betweenness_centrality(self.graph)
        plt.scatter(list(range(1,self.nodes+1)),[bc[i] for i in range(1,self.nodes+1)],c='r',s=[5])
        plt.xlabel("Nodes")
        plt.ylabel("Betweeness Centrality")
        plt.title("Betweeness Centrality of nodes")
        
    def closeness_vs_nodes(self):
        clc=nx.closeness_centrality(self.graph)
        plt.scatter(list(range(1,self.nodes+1)),[clc[i] for i in range(1,self.nodes+1)],c='r',s=[5])
        plt.xlabel("Nodes")
        plt.ylabel("Closeness Centrality")
        plt.title("Closeness Centrality of nodes")
    
    def betweeness_vs_closeness(self):
        bc=nx.betweenness_centrality(self.graph)
        clc=nx.closeness_centrality(self.graph)
        plt.scatter([bc[i] for i in range(1,self.nodes+1)],[clc[i] for i in range(1,self.nodes+1)],c='r',s=[5])
        plt.xlabel("Betweeness Centrality")
        plt.ylabel("Closeness Centrality")
        plt.title("Betweeness Centrality vs Closeness Centrality")
    
    def plot_C_vs_K(self):
        K=self.degree()
        C=self.clustering_coefficient()

        K=[K[i] for i in range(1,self.nodes+1)]
        C=[C[i] for i in range(1,self.nodes+1)]
        X=list(set(K))
        X.sort()
        Y=[]
        print("Unique values of Degree :",len(X))
        for i in X:
            temp=0
            c=0

            for j in range(len(K)):
                if K[j]==i:
                    temp+=C[j]
                    c+=1
            Y.append(temp/c)

        plt.scatter(X,Y,c='r',s=[5])
        plt.xlabel("Degree")
        plt.ylabel("Clustering Coefficient")
        plt.title("Clustering Coefficient vs Degree For Real Graph")
    
    def plot_knn_vs_K(self):
        d=self.knn()

        d=d.items()

        X=list(set([i[0] for i in d]))
        X.sort()
        Y=[]

        for i in X:
            temp=0
            c=0
            for j in d:
                if j[0]==i:
                    temp=temp+j[1]
                    c+=1

            Y.append(temp/c)

        plt.scatter(X,Y,c='r',s=[5])
        plt.xlabel("Degree")
        plt.ylabel("Average Connectivity(Knn)")
        plt.title("Average Connectivity( Knn ) vs Degree For Random Graph")
    
    def other_parameters(self):
        print("Number Of Nodes :",self.nodes)
        edges=len(self.graph.edges)
        print("Number Of Edges :",edges)
        print("Sparsity :", "{:.3f}".format(round(edges/((self.nodes*(self.nodes-1))/2),3)))
        print("Assortativity Coefficient :", "{:.3f}".format(round(nx.degree_assortativity_coefficient(self.graph),3)))
        print("Average Degree :","{:,.3f}".format(round(self.average_degree(),3)))
        
        
        
```

```python
US_network=Graph(graph)
```

```python
US_network.other_parameters()
```
